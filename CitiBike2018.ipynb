## Describe Data

```python
df201801 = pd.read_csv("Data/201801-citibike-tripdata.csv")
df201802 = pd.read_csv("Data/201802-citibike-tripdata.csv")
df201803 = pd.read_csv("Data/201803-citibike-tripdata.csv")
df201804 = pd.read_csv("Data/201804-citibike-tripdata.csv")
df201805 = pd.read_csv("Data/201805-citibike-tripdata.csv")
df201806 = pd.read_csv("Data/201806-citibike-tripdata.csv")
df201807 = pd.read_csv("Data/201807-citibike-tripdata.csv")
df201808 = pd.read_csv("Data/201808-citibike-tripdata.csv")
df201809 = pd.read_csv("Data/201809-citibike-tripdata.csv")
df201810 = pd.read_csv("Data/201810-citibike-tripdata.csv")
df201811 = pd.read_csv("Data/201811-citibike-tripdata.csv")
df201812 = pd.read_csv("Data/201812-citibike-tripdata.csv")
```

```python
df = pd.concat([df201801, df201802, df201803, df201804, \
                df201805, df201806, df201807, df201808, \
                df201809, df201810, df201811, df201812])
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
df.dtypes
```

### Convert starttime and stoptime to datetime

```python
df['starttime'] =  pd.to_datetime(df['starttime'])
df['stoptime'] =  pd.to_datetime(df['stoptime'])
```

```python
df.dtypes
```

```python
df.head(5)
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df['Start'] = pd.to_datetime(df['starttime']).dt.time
df['Stop'] = pd.to_datetime(df['stoptime']).dt.time
```

```python
df['month'] =  pd.to_datetime(df['starttime']).dt.month
df['day'] =  pd.to_datetime(df['starttime']).dt.day
```

```python
df['month_name'] = df['starttime'].dt.month_name()
df['day_name'] = df['starttime'].dt.day_name()
df.head(5)
```

## 1. How many trips have been recorded in total during 2018? 

```python
total_trips = df.tripduration.count()
print('The total number of trips in 2018 is:' + str(total_trips))
```

## 2. How many trips per month recorded in total during 2018? 

```python
trips_per_month = df.groupby(["month_name"])\
    ["tripduration"].count().sort_values(ascending=False).to_frame().reset_index()

trips_per_month.rename(columns={'tripduration':'No_of_Trips'}, inplace=True)
```

```python
trips_per_month
```

```python
trips_per_month.to_csv("trips_per_month.csv")
```

## 3. Monthly Trips by Gender 

```python
df['gender'].value_counts()
```

```python
gender_analysis = df[['month_name','gender','tripduration']]
```

```python
gender_analysis = gender_analysis[gender_analysis.gender != 0]
gender_analysis['gender'].replace(2, 'Female',inplace=True)
gender_analysis['gender'].replace(1, 'Male',inplace=True)
gender_analysis.head(5)
```

```python
gender_analysis['gender'].value_counts()
```

```python
MF_trips_per_month = gender_analysis.groupby(['month_name','gender'])\
    ['tripduration'].agg('count').to_frame().reset_index()

MF_trips_per_month.rename(columns={'tripduration':'No_of_Trips'}, inplace=True)
```

```python
MF_trips_per_month
```

```python
MF_trips_per_month.to_csv("trips_per_month_by_gender.csv")
```

## 4. Avg Trip time per month by Gender

```python
MF_mean_time_month = gender_analysis.groupby(['month_name','gender'])\
    ['tripduration'].agg(['mean','max','min']).astype(int).reset_index()

#MF_mean_time_month.rename(columns={'tripduration':'No_of_Trips'}, inplace=True)
```

```python
MF_mean_time_month
```

```python
MF_mean_time_month.to_csv("MF_mean_time_month.csv")
```

## 5.Top 10 Stations for start and end trips

```python
# Find top end stations
df_top_end_stations = df.groupby(["end station id"])\
    ["end station id"].count().sort_values(ascending=False)[0:10].to_frame()
df_top_end_stations = df_top_end_stations.rename(columns={"end station id":"no_of_times"})
df_top_end_stations
```

```python
# Find top start stations
df_top_start_stations = df.groupby(["start station id"])\
    ["start station id"].count().sort_values(ascending=False)[0:10].to_frame()
df_top_start_stations = df_top_start_stations.rename(columns={"start station id":"no_of_times"})
df_top_start_stations
```

```python
df_top_start_stations.to_csv("df_top_start_stations.csv")
df_top_end_stations.to_csv("df_top_end_stations.csv")
```
